Função em Python

```python
# Função para calcular a área de um círculo
def calcular_area_circulo(raio):
    pi = 3.14159
    return pi * raio ** 2

# Calculando a área para um raio de 5
area = calcular_area_circulo(5)
area
```

Visualização com Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Carregando um dataset de exemplo
iris = sns.load_dataset('iris')

# Criando um gráfico de dispersão com Seaborn
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)

# Exibindo o gráfico
plt.title("Gráfico de Dispersão: Comprimento vs Largura da Sépala")
plt.show()
```
